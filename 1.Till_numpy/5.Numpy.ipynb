```python
# scratch
def mean(data):
    return sum(data) / len(data)

def median(data):
    size = len(data)
    d = data.copy()
    d.sort()
    if size % 2 != 0:    
        mid = (size + 1) // 2
        return d[mid-1] # because in python we count from 0
    else:
        mid = size // 2
        return ( d[mid-1] + d[mid] ) / 2
    
```

```python
print(data.mean())
print(mean(data))
```

```python
print(np.median(data))
print(median(data))
```

```python
def plot_data(data, ylims=None):
    m = data.mean()
    mu = np.median(data)
    print("Mean: ", m)
    plt.figure(dpi=100)
    plt.plot(range(len(data)), data, 'ro')
    plt.hlines(m, 0, len(data), lw=2, color='black', label=f'Mean = {m}')
    plt.hlines(mu, 0, len(data), lw=2, color='magenta', label=f'Median = {mu}')
    for x, y2 in enumerate(data):
        plt.vlines(x, m, y2, color='c', ls=':')
    plt.legend()
    if ylims is not None:
        plt.ylim(ylims)
    plt.show()
```

```python
plot_data(data)
# mean == median -> there is no outlier in data
# data is symmetric
```

```python
plot_data(data_1)
```

```python
plot_data(data_1, [0, 50])
```

```python
data_2 = data_1.copy()
data_2 = np.append(data_1, [200, 150])
```

```python
plot_data(data_2, [0, 50])
```

    mean is sensitive towards outliers thats why we rarely use mean

    median is a good measure of center just because it always points to center and do not get effected by outliers


    We can easily check outliers by comparing mean and median 
    
    
        mean === median  -> Data is Symeetric - no outliers
        
        mean >>> median  -> Data is Right Skewed - positive / high outliers
        
        mean <<< median -> Data is Left Skewed - negative / low outliers
