```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u = np.random.rand(100000)
plt.hist(u, bins=100)
plt.xlabel("Data", fontsize=20)
plt.ylabel("Frequency", fontsize=20)
plt.title("Uniform Distribution", color='red', fontsize=25)
plt.show()
```

```python
n = np.random.randn(100000)
plt.hist(n, bins=100)
plt.title("Normal Distribution", color='salmon', fontsize=25)
plt.xlabel("Data", fontsize=20)
plt.ylabel("Frequency", fontsize=20)
plt.show()
```

```python

plt.hist(u, bins=100, histtype='step', color='blue', label='rand')
plt.hist(n, bins=100, histtype='step', color='red', label='randn')
plt.title("Normal Distribution", color='salmon', fontsize=25)
plt.xlabel("Data", fontsize=20)
plt.ylabel("Frequency", fontsize=20)
plt.show()
```

```python

plt.hist(u, bins=100, histtype='step', color='blue', label='rand', density=True)
plt.hist(n, bins=100, histtype='step', color='red', label='randn', density=True)
plt.title("Normal Distribution", color='salmon', fontsize=25)
plt.xlabel("Data", fontsize=20)
plt.ylabel("Density", fontsize=20)
plt.show()
```

### First Rule 

    1. they are equal, or

    2. one of them is 1

    Image  (3d array): 256 x 256 x 3
    Scale  (1d array):             3
    Result (3d array): 256 x 256 x 3

    A      (4d array):  8 x 1 x 6 x 1
    B      (3d array):      7 x 1 x 5
    Result (4d array):  8 x 7 x 6 x 5

    A      (2d array):  5 x 4
    B      (1d array):      1
    Result (2d array):  5 x 4

    A      (2d array):  5 x 4
    B      (1d array):      4
    Result (2d array):  5 x 4

    A      (3d array):  15 x 3 x 5
    B      (3d array):  15 x 1 x 5
    Result (3d array):  15 x 3 x 5

    A      (3d array):  15 x 3 x 5
    B      (2d array):       3 x 5
    Result (3d array):  15 x 3 x 5

    A      (3d array):  15 x 3 x 5
    B      (2d array):       3 x 1
    Result (3d array):  15 x 3 x 5
