#### Apply - Map

```python
df2 =  pd.DataFrame({
    'height': np.random.randint(120, 160, 10),
    'weight': np.random.randint(50, 120, 10),
    'salary': np.random.randint(30, 80, 10),
    'country': np.random.choice(['India', 'USA', 'AUS'], 10, replace=True)
}, index=[ chr(65+i) for i in range(10)])
```

    60-100 

    60 -> outlier
    100 -> outler

    replace outliers with median

```python
df2.loc[df2[ (df2['weight'] < 60) | (df2['weight'] > 100) ].index, 'weight'] # = replacement value
```

```python
df2['weight']
```

```python
median = df2['weight'].median()
median
```

```python
def function(value):
    if value > 100 or value < 60:
        return median
    return value
df2['weight'].apply(function)
```

```python
df2['sex'] = pd.Series(np.random.choice(["Male", "Female"], 10, replace=True),
                       index=df2.index)
```

```python
df2 = df2[['sex', 'country', 'height', 'weight', 'salary']]
```

```python
df2
```

```python
df2.groupby('sex').agg({
    'height': 'mean',
    'weight': 'mean',
    'salary': 'mean',
    'country': 'count'
})

# male height, weight, salary > female => Select
# Biased 
```
