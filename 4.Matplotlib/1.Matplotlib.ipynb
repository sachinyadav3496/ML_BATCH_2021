```python
x = np.arange(10)
y1 = y
y2 = y + np.random.random()*5
y3 = y - np.random.random()*5
l1, l2, l3 = plt.plot(x, y1, x, y2, x, y3)
l1.set_color('salmon')
l2.set_color('cyan')
l3.set_color('blue')
plt.axis([-4, 15, 0, 55])
#        xmin, xmax, ymin, ymax
```

```python
fig = plt.figure(facecolor='salmon')

ax1 = fig.add_axes( [0, 0, 1, 1] )

ax2 = fig.add_axes( [0.45, 0.05, 0.5, 0.4])

x = np.linspace(-5, 5, 100)
ax1.plot(x, x**2, label='square')
ax1.plot(x, x**3, label='cube')
ax1.legend()

a = np.linspace(-6.28, 6.28, 120)
ax2.plot(a, np.sin(a), label='sin')
ax2.plot(a, np.cos(a), label='cos')
ax2.legend()

plt.show()
```

```python
fig = plt.figure(facecolor='salmon')

ax1 = fig.add_axes( [0, 0, 1, 0.4] )

ax2 = fig.add_axes( [0, 0.5, 0.5, 0.4])

x = np.linspace(-5, 5, 100)
ax1.plot(x, x**2, label='square')
ax1.plot(x, x**3, label='cube')
ax1.legend()

a = np.linspace(-6.28, 6.28, 120)
ax2.plot(a, np.sin(a), label='sin')
ax2.plot(a, np.cos(a), label='cos')
ax2.legend()

plt.show()
```

```python
font_dict = { 'color': 'white',  'fontfamily': 'monospace', 'fontweight':  'extra bold'}
```

```python
fig = plt.figure(facecolor='red')
ax = fig.add_axes([0, 0, 1, 1], facecolor='black')
x = np.linspace(-5, 5, 10)

ax.plot(x, x**2, 'r--o', label='square')
ax.plot(x, x**3, 'w:s', label='cube')

ax.legend()
ax.set_title("Anatomy of figure", fontsize=30, fontdict=font_dict)
ax.set_xlabel("x Values", fontsize=20, fontdict=font_dict)
ax.set_ylabel("y Values", fontsize=20, fontdict=font_dict)

ax.set_xticks(np.arange(-6, 6))
ax.set_xticklabels([chr(65+i) for i in range(12)], rotation=45, fontdict=font_dict)
ax.set_yticks(np.arange(-130, 130, 20))

ax.axis([-8, 8, -150, 150])
#            xmin, xman, ymin, ymax

ax.grid()
plt.show()
```
