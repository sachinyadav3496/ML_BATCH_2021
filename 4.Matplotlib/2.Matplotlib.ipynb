```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 100)
```

```python
fig1 = plt.figure(num=1)
fig2 = plt.figure(num=2)
ax1 = fig1.add_axes([0, 0, 1, 1])
ax2 = fig2.add_axes([0, 0, 1, 1])

for sp in ['right', 'left', 'top', 'bottom']:
    ax1.spines[sp].set_visible(False)

ax1.plot(x, x**2)
ax2.plot(x, x**3)
plt.show()
```

```python
plt.plot(x, x**2) # figure, ax, plot
plt.plot(x, x**3)

ax = plt.gca() # get current axes
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_color('red')
ax.spines['left'].set_color('blue')
ax.spines['left'].set_lw(2)
ax.spines['bottom'].set_lw(2)
ax.spines['bottom'].set_ls('--')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 5), dpi=100)

ax1 = fig.add_subplot(3, 3, 1)
ax5 = fig.add_subplot(3, 3, 5)
ax9 = fig.add_subplot(3, 3, 9)

ax1.plot(x, x**2)
ax5.plot(x, x**3)
ax9.plot(x, np.sin(x))
plt.tight_layout()
plt.savefig('one.png', transparent=True, )
plt.show()
```

## Scatter Plot 

    Relationship Plot

```python
height = np.random.normal(140, 20, 100)
weight = np.random.normal(20, 5, 100) + 0.4*height
```

```python
plt.figure(figsize=(8, 4), dpi=100)

plt.title('Height vs Weight')
plt.xlabel("Height")
plt.ylabel('Weight')



plt.scatter(height, weight, color='#cccccc')

bx = height > 165
plt.scatter(height[bx], weight[bx], c='red')

bx = height < 120
plt.scatter(height[bx], weight[bx], c='green')
plt.show()
```

```python
colors = [ 'red', 'green', 'blue', 'magenta', 'cyan', 'yellow', 'salmon', 'purple', 'orange']

sizes = np.arange(10, 150, 40)
cls = np.random.choice(colors, 100, replace=True)
szs = np.random.choice(sizes, 100, replace=True)
plt.figure(figsize=(8, 4), dpi=100)

plt.title('Height vs Weight')
plt.xlabel("Height")
plt.ylabel('Weight')



plt.scatter(height, weight, c=cls, s=szs, marker='h')
plt.show()
```
