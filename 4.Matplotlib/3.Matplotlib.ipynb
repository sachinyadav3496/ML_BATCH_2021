```python
fig = plt.figure(dpi=100)
ax = fig.add_axes([0, 0, 1, 1])
freq, bins, patches = ax.hist(height, bins=30, ec='k')

tks, tk_lbs = plt.xticks(bins.astype(int), rotation=90)
plt.yticks(np.arange(0, 91, 5))
#print(freq)
#print(bins)
ac = [ str(i) for i in range(10) ] + [ chr(65+i) for i in range(6)] * 4

rm_color = lambda : "#"+"".join(list(np.random.choice(ac, 6)))

wd = patches[0].get_width() // 2
for p, t in zip(patches, tk_lbs):
    
    color = rm_color()
    p.set_edgecolor('#333333')
    p.set_facecolor(color)
    t.set_color(color)
    x, y = p.get_x()+wd, p.get_height()+wd
    plt.text(x, y, int(p.get_height()), color='k', rotation=90, ha='center')
for sp in ['right', 'top']:
    ax.spines[sp].set_visible(False)
plt.grid(ls=':', alpha=0.3)


plt.text(80, 90, "Sachin", color='red', fontsize=20, horizontalalignment='center', rotation=0)
plt.show()
```

```python
lb = height.mean() - height.std()
up = height.mean() + height.std()
```

```python
lb
```

```python
up
```

```python
def in_bound(value):
    if value >= lb and value <= up:
        return True
    return False
```

```python
fig = plt.figure(dpi=100)
ax = fig.add_axes([0, 0, 1, 1])
freq, bins, patches = ax.hist(height, bins=30, ec='k')

tks, tk_lbs = plt.xticks(bins.astype(int), rotation=90)
plt.yticks(np.arange(0, 91, 5))
#print(freq)
#print(bins)
ac = [ str(i) for i in range(10) ] + [ chr(65+i) for i in range(6)] * 4

rm_color = lambda : "#"+"".join(list(np.random.choice(ac, 6)))

wd = patches[0].get_width() // 2
for p, t in zip(patches, tk_lbs):
    
    color = rm_color()
    p.set_edgecolor('#333333')
    if in_bound(p.get_x()):
        p.set_facecolor('salmon')
    else:
        p.set_facecolor('#cccccc')
    t.set_color(color)
    x, y = p.get_x()+wd, p.get_height()+wd
    plt.text(x, y, int(p.get_height()), color='k', rotation=90, ha='center')
for sp in ['right', 'top']:
    ax.spines[sp].set_visible(False)
plt.grid(ls=':', alpha=0.3)


plt.text(80, 90, "Sachin", color='red', fontsize=20, horizontalalignment='center', rotation=0)
plt.show()
```

##### To check Data Distribution we use Histograms

```python
obs = plt.boxplot(height, showmeans=True)

for ob in obs['whiskers']:
    ob.set_color('red')
    ob.set_lw(3)
    ob.set_ls('--')
for cap in obs['caps']:
    cap.set_color('purple')
    cap.set_lw(2)
    
for f in obs['fliers']:
    f.set_markeredgecolor('yellow')
    f.set_markerfacecolor('pink')
    f.set_markersize(10)
    #f.set_edgecolor('red')
for m in obs['medians']:
    m.set_color('green')
    m.set_lw(3)
    
for m in obs['means']:
    m.set_markerfacecolor('white')
    m.set_markeredgecolor('yellow')
```
