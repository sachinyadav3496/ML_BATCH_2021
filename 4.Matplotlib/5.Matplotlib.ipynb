## Plots with Matplot

### Annotation

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(-5, 5, 100)

y = x ** 2
```

```python
plt.figure(dpi=100)
plt.plot(x, y, 'b-..')
px = 2
py = 4

plt.plot(px, py, 'or')

plt.text(-1, 4, 'Important Point', ha='left', fontsize=10, color='red')
plt.text(px, py, f"x = {px:0.2f}\ny = {py:0.2f}", rotation=47, color='gray')
plt.text(-2, 15, f"Square Function\n\t$y = x^2$", color='blue', fontsize=20)
plt.show()
```

```python
plt.figure(dpi=100)
plt.plot(x, y, 'b-.')
px = 2
py = 4

plt.plot(px, py, 'or')

bbox = { 'boxstyle': 'rarrow,pad=0.3', 'ec': 'k', 'lw':2, 'fc':'lightgray'}
bboxs = { 'boxstyle': 'square,pad=0.3', 'ec': 'r', 'lw':2, 'fc':'lightblue'}

plt.text(-2, 15, f"Square Function\n\t$y = x^2$", color='blue', fontsize=20,
        bbox=bboxs)

plt.annotate("Important Point", xy=(px, py), xytext=(px-3.5, py),
            color='red', fontsize=10, weight='heavy',bbox=bbox,
             )

plt.show()
```

#### xKCD plots 

```python
with plt.xkcd():
    plt.figure(dpi=100)
    plt.plot(x, y, 'b-.')
    px = 2
    py = 4

    plt.plot(px, py, 'or')

    bbox = { 'boxstyle': 'rarrow,pad=0.3', 'ec': 'k', 'lw':2, 'fc':'lightgray'}
    bboxs = { 'boxstyle': 'square,pad=0.3', 'ec': 'r', 'lw':2, 'fc':'lightblue'}

    plt.text(-2, 15, f"Square Function\n\t$y = x^2$", color='blue', fontsize=20,
            bbox=bboxs)

    plt.annotate("Important Point", xy=(px, py), xytext=(px-3.5, py),
                color='red', fontsize=10, weight='heavy',bbox=bbox,
                 )

    plt.show()
```

```python
x = np.linspace(-6, 6)
y = np.linspace(-6, 6)

xx, yy = np.meshgrid(x, y)

def function(xx, yy):
    return np.sin(xx + yy)


zz = function(xx, yy)

%matplotlib notebook
fig = plt.figure(dpi=100)

ax = fig.add_axes([0, 0, 1, 1], projection='3d')
ax.scatter(xx, yy, zz)


ax.set_xlabel("x")
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

x = np.linspace(-6, 6)
y = np.linspace(-6, 6)



%matplotlib notebook
fig = plt.figure(dpi=100)

ax = fig.add_axes([0, 0, 1, 1],  projection='3d')

ax.plot_surface(X, Y, Z, cmap='coolwarm') # cool - warm , low-high, blue - red
ax.grid()

ax.set_xlabel("x")
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

x = np.linspace(-6, 6)
y = np.linspace(-6, 6)



%matplotlib inline
fig = plt.figure(dpi=100)

ax = fig.add_axes([0, 0, 1, 1],  projection='3d')

ax.contourf(X, Y, Z, cmap='coolwarm') # cool - warm , low-high, blue - red
plt.colorbar()
ax.grid()

ax.set_xlabel("x")
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
